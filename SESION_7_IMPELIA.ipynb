- Ordenar la serie por edad en orden ascendente.
- Ordenar la serie por nombre en orden descendente.
- Mostrar solo los nombres de personas cuyo nombre comience con la letra 'M'.
- Calcular la media de las edades de los mayores de 20


```python
import pandas as pd
import numpy as np
```

```python
# Define a list of possible first names
first_names = ["Alice", "Bob", "Charlie", "Dave", "Eve", "Jona", "Katie", "Liam", "Mia", "Noah", "Olivia", "Peter", "Quinn", "Ryan", "Sophia", "Thomas", "Ursula", "Vincent", "Wendy", "Xavier"]
```

```python
# Generate 20 random ages between 18 and 65
ages = np.random.randint(18, 66, size=20)
```

```python
# Create a dictionary with first names as keys and ages as values
name_age_dict = {name: age for name, age in zip(first_names, ages)}
```

```python
name_age_dict
```

```python
name_age_serie = pd.Series(name_age_dict)
```

```python
name_age_serie
```

```python
name_age_serie[name_age_serie>20]
```

```python
name_age_serie.sort_values()
```

```python
name_age_serie.sort_values(ascending=False)
```

```python
name_age_serie[name_age_serie.index.str.startswith("M")]
```

```python
name_age_serie[name_age_serie>20].mean()
```

Tareas:

- Concatenar los DataFrames: Concatenar los DataFrames de enero y febrero en un √∫nico DataFrame.

- Filtrar los datos: Filtrar las filas en las que las ventas sean mayores a 5000.

- Agrupar por y agregar: Agrupar los datos filtrados por regi√≥n y categor√≠a, y calcular las ventas totales por grupo.

- Ordenar los datos: Ordenar el DataFrame resultante por ventas en orden descendente.

```python
# Datos de ventas de enero
data_january = {
'Regi√≥n': ['Norte', 'Norte', 'Este', 'Este', 'Sur', 'Sur', 'Oeste', 'Oeste'],
'Categor√≠a': ['Electr√≥nica', 'Muebles', 'Electr√≥nica', 'Muebles', 'Electr√≥nica', 'Muebles', 'Electr√≥nica', 'Muebles'],
'Ventas': [5000, 7000, 3000, 10000, 2000, 1500, 6000, 8000]
}
df_january = pd.DataFrame(data_january)
```

```python
# Datos de ventas de febrero
data_february = {
'Regi√≥n': ['Norte', 'Norte', 'Este', 'Este', 'Sur', 'Sur', 'Oeste', 'Oeste'],
'Categor√≠a': ['Electr√≥nica', 'Muebles', 'Electr√≥nica', 'Muebles', 'Electr√≥nica', 'Muebles', 'Electr√≥nica', 'Muebles'],
'Ventas': [4000, 6000, 2000, 12000, 2500, 1700, 5500, 9000]
}
df_february = pd.DataFrame(data_february)
```

```python
df_january
```

```python
df_february
```

```python
df_concat = pd.concat([df_january,df_february])
```

```python
df_concat
```

```python
df_concat[df_concat["Ventas"]>5000]
```

```python
df_concat.groupby(["Regi√≥n","Categor√≠a"])["Ventas"].sum()
```

```python
df_concat.groupby(["Regi√≥n","Categor√≠a"])["Ventas"].sum().sort_values(ascending=False)
```

Recuento de personas que vivieron por clase

```python
df=pd.read_csv('https://raw.githubusercontent.com/jorloicono/AF-PYTHON/refs/heads/main/DATOS/titanic.csv')
```

```python
df
```

```python
df.groupby("Pclass")["PassengerId"].count().sort_values(ascending=False)
```

```python
ax = df.groupby("Pclass")["PassengerId"].count().sort_values(ascending=False).plot(kind="bar")
total = len(df)
for p in ax.patches:
    percentage = '{:.1f}%'.format(100 * p.get_height()/total)
    x = p.get_x() + p.get_width() / 2 - 0.1
    y = p.get_y() + p.get_height()
    ax.annotate(f'{p.get_height()}\n({percentage})', (x, y), size = 12)
```

```python
df[df['Survived']==1].groupby("Pclass")["PassengerId"].count()
```

```python
df.groupby(["Pclass", "Survived"]).size().unstack()
```

```python
df.groupby(["Pclass", "Survived"]).size().unstack().plot(kind = "bar", stacked = True, title = "N√∫mero de personas fallecidas y supervivientes por clase")
```
